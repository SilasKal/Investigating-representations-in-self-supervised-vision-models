# 1. Render Images Using MAPS

```python
import numpy as np

from maps.dataset import Dataset
```

```python
object_name = 'white-shark_001'
root = '/work/dldevel/galella/datasets/MAPS/scenes'
```

```python
dataset = Dataset(root=root)
print(dataset)
```

```python
scene = dataset.get_scene(object_name)
print(scene)
```

```python
params = scene.get()

for param, value in params.items():
    print(f'{param}: {value}')
```

```python
params['camera.azimuth']
```

```python
img = scene.render()
img
```

```python
params.update({
    'camera.distance': 1.5,    
    'camera.roll': np.pi / 4,
    'camera.elevation': np.pi / 2,
    'light.power': 0.4,
    'background.noise': 0.2,
})

scene.set(params)
scene.render()
```
